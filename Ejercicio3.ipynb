Pregunta 1: ¿Qué R2 se obtiene si se calcula una regresión lineal entre la duración del vuelo y su precio? 

Ayuda: 

Con statsmodels puede calcularse así: 
```

```

Con sklearn puede calcularse así: 
```

X=['duration']

```
